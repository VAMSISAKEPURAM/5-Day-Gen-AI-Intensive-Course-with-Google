### Genaration parameters 
* generation parameters are the settings you pass to a language model (like Groq, Gemini, or OpenAI models) that control how the text is produced. Think of them as the “knobs” you can adjust to influence the style, length, and randomness of the output
* * <b>   temperature=0.2,</b>          # Lower randomness for focused output
  *    max_tokens=150,           # Shorter response length
  *    top_p=0.9,                # Consider a wide but reasonable set of words
  *    frequency_penalty=0.5,    # Lightly discourage word repetition
  *    presence_penalty=0.3,     # Lightly encourage new concepts
  *    seed=42                   # For reproducible results
)

### Prompting 

## Technique	What It Does	When to Use	Example Prompt Snippet
* Zero-Shot	The model gets no example; just an instruction.	Simple, well-known tasks (translation, summarization).	
"Translate 'Hello' to Spanish."
* Few-Shot	Provide a few input-output examples before asking the real question.	Tasks with specific formatting or style needs.	"English: cat → French: chat\nEnglish: dog → French: chien\nEnglish: house → French:"
* Chain-of-Thought (CoT)	Ask the model to reason step-by-step before giving a final answer.	Math problems, logic, multi-step reasoning.	"Roger has 5 balls. He buys 2 packs of 3 balls each. How many now? Let's think step by step."
* Self-Consistency	Run CoT multiple times (with temperature > 0) and pick the most common answer.	When accuracy on reasoning is critical (e.g., arithmetic).	(Implemented in code, not a single prompt.)
* Tree-of-Thoughts (ToT)	Explore multiple reasoning paths at each step, evaluate them, and choose the best.	Complex problems requiring planning or search.	(Advanced, usually requires multiple API calls.)
Prompt Chaining	Break a task into sub-tasks, each handled by a separate prompt, piping outputs.	Long documents, multi-step transformations.	Prompt 1: "Summarize this article." Prompt 2: "Extract key dates from summary."
* Role Prompting	Assign a persona or role to the LLM.	Stylized answers, expert advice, code generation.	"You are a senior Python developer. Write a function to sort a list."
* Negative Prompting	Explicitly tell the model what not to do.	Avoid hallucinations, off-topic responses, or bad formatting.	"Do not mention any other countries. Only talk about France."
* Structured Output (JSON/XML)	Request output in a machine‑readable format.	Downstream processing, API integration.	"Output a JSON object with keys: name, age, city."
* System Prompt	Set persistent instructions for the whole conversation (available in Groq API).	Setting tone, rules, or context for a multi‑turn chat.

````python

zero_shot_prompt = """Classify movie reviews as POSITIVE, NEUTRAL or NEGATIVE.
Review: "Her" is a disturbing study not revealing the direction
humanity is headed if AI is not  allowed to keep evolving,
checked. I wish not there were more movies like this worest movie
.
Sentiment: """

response = client.chat.completions.create(
    model = 'llama-3.3-70b-versatile',
    messages = [
        {
            'role': 'user',
            'content': zero_shot_prompt
        }
    ],

    max_tokens = 100,
    temperature=0.2,
    top_p=0.9,
    
)
print(response.choices[0].message.content)
````
